### Definicin del Modelo

Se define un modelo de regresin logstica el cual toma como entrada las imagenes de entrenamiento y calcula los logits de las categoras de clasificacin. Posterior al clculo de logits, se realizar una entropa cruzada entre los logits y la matrz de one hot encoding de las categoras, retornando como resultado una matriz de vectores con pesos calculados por cada categora.

Este modelo de regresin funciona con mini batch gradient descent, con el batches de tamao 32.

```python
def logistic_regression(x, y , epochs, lr, batch_size, print_interval,logs_dir):
    tf.reset_default_graph()
    predicted_weights = None
    
    total_iterations = int(x.shape[0]/batch_size)

    with tf.name_scope("Pesos"):
        x_train = tf.placeholder(tf.float32,[batch_size,None],"X")
        train_labels = tf.placeholder(tf.float32,[batch_size,y.shape[1]],"Train_Labels")
        weights = tf.Variable(name="Weights", initial_value=tf.truncated_normal(([x.shape[1],y.shape[1]]),dtype=tf.float32))
        biases = tf.Variable(tf.zeros(y.shape[1]))
    with tf.name_scope("Calcular_Logits"):
        logits = tf.matmul(x_train,weights) + biases
    with tf.name_scope("Calcular_Costo"):
        loss = tf.reduce_mean(tf.nn.softmax_cross_entropy_with_logits(logits=logits,labels=train_labels))
    with tf.name_scope("Calcular_Gradientes"):
        gradients = tf.gradients(loss ,weights)
    with tf.name_scope("Actualizar_Pesos"):
        update_weights = tf.assign(weights,(tf.add(weights,tf.scalar_mul(-lr,gradients[0]))))

    loss_summary = tf.summary.scalar(name='Loss_vs_Epochs', tensor=loss)

    with tf.Session() as session:
        curr_datetime = datetime.datetime.now().strftime('%Y%m%d_%H%M%S')
        writer = tf.summary.FileWriter(logs_dir+curr_datetime+"_lr="+str(lr)+"_epochs="+str(epochs),session.graph)
        session.run(tf.global_variables_initializer())
        x_mb = None
        y_mb = None
        for epoch in range(epochs):
            for i in range(total_iterations):
                batch_start  = i*batch_size
                batch_end = batch_start + batch_size
                
                x_mb = np.array(x[batch_start:batch_end])
                y_mb = np.array(y[batch_start:batch_end])
                
                session.run(update_weights, feed_dict={x_train: x_mb, train_labels: y_mb})
                
            if ((epoch+1) % print_interval) == 0:
                print("epoch:"+str(epoch+1)+" costo: "+str(session.run(loss, feed_dict={x_train: x_mb, train_labels: y_mb})))
      
            writer.add_summary(session.run(loss_summary, feed_dict={x_train: x_mb, train_labels: y_mb}), epoch)
        

        predicted_weights = session.run(weights)
        writer.close()
    session.close()
    
    return predicted_weights
```

### Prediccin de clases
Se crea una funcin que recibe como entrada una imagen y los pesos calculados durante el entrenamiento, devolviendo como resultado la distribucin de probabilidad a la que pertenece dicha imagen.

```python
def predict_class(image,weights):
    tf.reset_default_graph()
    imageph = tf.placeholder(tf.float32,[image.shape[0],1],"Predicted_Image")
    with tf.Session() as session:
        prediction = session.run(tf.nn.softmax(tf.matmul(tf.transpose(imageph),weights)),feed_dict={imageph: image})
    session.close()
    return prediction
```

```python
def predict_image(image,weights,label,encoded_labels,classes,scale=1):
    plt.figure(figsize=(2,2))
    plt.imshow(image.reshape(28,28), cmap="gray")
    plt.show()
    
    predicted_distribution = predict_class(image.reshape(-1,1)/scale,weights)
    print("Distribucin de Probabilidad:")
    print(predicted_distribution)
    print("\nClase Real: "+label+", Prediccin: "+classes[encoded_labels[np.argmax(predicted_distribution)]])
    print("----------------------------------------------------------\n")
    
```

### Filtrado de imagenes y One-hot encoding

```python
#diccionario de clases
classes = {
  0: "T-shirt/top",
  5: "Sandal",
  7: "Sneaker"
}

#filtrado set de entrenamiento
image_size = test_images[0].shape[0] * test_images[0].shape[1]

train_mask = np.isin(train_labels, list(classes.keys()))
filtered_train_images = (train_images[train_mask]).reshape(-1,image_size)
filtered_train_labels = train_labels[train_mask]

#filtrado set de pruebas
test_mask = np.isin(test_labels, list(classes.keys()))
unshaped_test_images = test_images[test_mask]
filtered_test_images = unshaped_test_images.reshape(-1,image_size)
filtered_test_labels = test_labels[test_mask]

#one hot encoding
train_one_hot_encoded = pd.get_dummies(filtered_train_labels,drop_first=False)

print("One-hot encoding de clases seleccionadas")
train_one_hot_encoded.head(10)
```

### Entrenamiento y prediccin del modelo

```python
weights = logistic_regression(x=filtered_train_images,y=train_one_hot_encoded.to_numpy(),epochs=1000, lr=0.001, batch_size=32, print_interval=100,  logs_dir="./fashion_mnist/")
```

<img src=images/loss1.png />

De los resultados que se obtienen de costo durante el entrenamiento, se puede observar que algo no va bien y se confirma con la grfica de costo generada en tensorboard, el costo a lo largo de los epochs se ve bastante inestable.

Se proceder a realizar prediccin con los resultados obtenidos y validar los resultados.

```python
for i in range(5):
    index = random.randint(0,len(filtered_test_images))
    image = filtered_test_images[index]
    label = classes[filtered_test_labels[index]]
    labels_encoded = train_one_hot_encoded.columns.values.tolist()
    predict_image(image, weights,label,labels_encoded,classes)
```

### Escalado de datos

A pesar que se han logrado predecir de manera correcta las clases de prenda al ingresar los valores de las imagenes sin escalar, se puede observar que el modelo de prediccin devuelve el one hot encoding como resultado en lugar de una distribucin de probabilidad, lo cual no es el resultado esperado al aplicar softmax a los logits.

Se proceder a realizar un scaling de los datos, dividiendo los valores de las imagenes de entrenamiento por el valor de 255 ya que la escala de grises es de 0 a 255.

```python
scaled_weights = logistic_regression(x=filtered_train_images/255,y=train_one_hot_encoded.to_numpy(),epochs=1000, lr=0.001, batch_size=32, print_interval=100,  logs_dir="./fashion_mnist/")
```

<img src=images/loss2.png />

En este caso se puede ver que la funcin de costo converge de una manera estable. Se procede a realizar las predicciones de 10 imagenes tomadas aleatoriamente del dataset de pruebas.

```python
for i in range(5):
    index = random.randint(0,len(filtered_test_images))
    image = filtered_test_images[index]
    label = classes[filtered_test_labels[index]]
    labels_encoded = train_one_hot_encoded.columns.values.tolist()
    predict_image(image, scaled_weights,label,labels_encoded,classes,255)
```

Ahora podemos validar que al escalar los datos, las predicciones han sido acertadas y el resultado obtenido por softmax es una distribucin de probabilidad entre las 3 clases entrenadas.

### Conclusiones

En este caso en particular, al no escalar los datos, se lograba predecir la clase de la prenda de manera adecuada, sin embargo, la distribucin de probabilidad obtenida por el modelo no era la correcta.
